# YouTube Adverts Twitter Scrape

## General YouTube search

```python
c = twint.Config()
c.Search = "youtube ads"
c.Format = "Tweet: {tweet} | Date: {date}"
c.Since = "2020-03-01"
c.Until = "2020-04-27"
c.Limit = 1500
c.Lang = "en"
c.Count = True
c.Pandas = True
c.Store_csv = True
c.Output = "youtubeadstweets.csv"
#configures twint to search twitter for the term 'youtube ads'
#with the following parameters, and saves it to a csv file named 'youtubeadstweets.csv'

twint.run.Search(c)
#executes search
```

```python
c = twint.Config()
c.Search = "youtube adverts"
c.Format = "Tweet: {tweet} | Date: {date}"
c.Since = "2020-03-01"
c.Until = "2020-04-27"
c.Limit = 1500
c.Lang = "en"
c.Count = True
c.Pandas = True
c.Store_csv = True
c.Output = "youtubeadstweets.csv"
#tweaked search term to 'adverts', writes tweets to same csv file as above

twint.run.Search(c)
```
